## Import Libraries & Data

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Import dataframe
path = r'C:\\Users\\sstorer\\OneDrive\\Instacart Basket Analysis'
ords_prods_cust = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_customers.pkl'))
```

```python
# Check the shape of dataframe
ords_prods_cust.shape
```

## Create a histogram of the “order_hour_of_day” column

```python
# Create a histogram showing the frequency of the order_hour_of_day column
hist_hod = ords_prods_cust['order_hour_of_day'].plot.hist(bins = 24)
```

This histogram shows the frequency of orders made at each hour of the day. The busiest hours of the day are from around 8am-4pm.

## Create a bar chart from the “loyalty_flag” column

```python
# Create a bar chart based on the loyalty_flag column
bar_loyalty_flag = ords_prods_cust['loyalty_flag'].value_counts().plot.bar()
```

## Check whether there’s a difference in expenditure depending on the hour of the day

```python
# Create a random sample of the dataframe
np.random.seed(4)
dev = np.random.rand(len(ords_prods_cust)) <= 0.7
```

```python
# Split the dataframe into the two samples
big = ords_prods_cust[dev]
small = ords_prods_cust[~dev]
```

```python
# Narrow the df down to the necessary columns
df_sample = small[['prices', 'order_hour_of_day']]
```

```python
# Create a line chart showing price by hour of the day
line_price_hod = sns.lineplot(data = df_sample, x = 'order_hour_of_day',y = 'prices')
```

## Determine whether there’s a connection between age and family situation

```python
# Create a line chart exploring the connections between age and number of dependents
line_age_dependents = sns.lineplot(data = small, x = 'age',y = 'number_of_dependents')
```

There is no correlation between age and number of dependents.

## Explore whether there’s a connection between age and spending power

```python
# Create a scatter plot to show the relationship between age and income
scatter_age_income = sns.scatterplot(x = 'age', y = 'income',data = small)
```

There is an increase in income after age 40.

## Export Visualizations

```python
# Export images
hist_hod.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_order_hour_of_day.png'))
bar_loyalty_flag.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_loyalty_flag.png'))
line_price_hod.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_price_hod.png'))
line_age_dependents.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_age_dependents.png'))
scatter_age_income.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scatter_age_income.png'))
```
