## Import Libraries and Data

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Import dataframe
path = r'C:\\Users\\sstorer\\OneDrive\\Instacart Basket Analysis'
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_grouped.pkl'))
```

## Create Bar Charts

```python
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
bar = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# Export bar chart
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

## Create Histograms and Scatterplots

```python
# Create a histogram
ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# Create a scatterplot
sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge)
```

```python
hist = ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# Export histogram
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices.png'))
```

## Sample Data

```python
# Create a random sample of the dataframe
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Split the dataframe into the two samples
big = ords_prods_merge[dev]
small = ords_prods_merge[~dev]
```

```python
# Check the length of the dataframes
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

## Create Line Chart

```python
# Narrow down to only the necessary columns
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Create a line graph
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export line chart
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_prices_dow.png'))
```
